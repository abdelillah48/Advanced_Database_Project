```python
import pandas as pd

dataset_path = "/content/drive/MyDrive/airbnb-listings@public.csv"
df = pd.read_csv(dataset_path, sep=';', low_memory=False)
df.head(4)
```

```python
print(df['Country Code'])
```

```python
print(df.columns)
```

```python
colonnes_a_garder = ['ID', 'Host ID', 'Host Name', 'Host Location', 'Host Response Time','Host Response Rate', 'Street', 'City', 'State', 'Country Code','Country', 'Room Type', 'Accommodates', 'Bathrooms', 'Bedrooms', 'Beds','Bed Type', 'Price', 'Weekly Price', 'Monthly Price', 'Minimum Nights','Maximum Nights', 'Calendar Updated', 'Number of Reviews','First Review', 'Last Review', 'Review Scores Rating','Review Scores Accuracy', 'Review Scores Cleanliness','Review Scores Checkin', 'Review Scores Communication','Review Scores Location', 'Review Scores Value', 'Reviews per Month']

# Supprimer les colonnes qui ne sont pas dans la liste colonnes_a_garder
colonnes_a_supprimer = [colonne for colonne in df.columns if colonne not in colonnes_a_garder]
df = df.drop(colonnes_a_supprimer, axis=1)

# Supprimer les lignes avec des valeurs manquantes
df = df.dropna()

# Sauvegarder le nouveau dataset nettoyé
nouveau_chemin = "/content/drive/MyDrive/Airbnb_Data_Nouveau.csv"
df.to_csv(nouveau_chemin, index=False)  # L'argument index=False évite d'écrire l'index dans le fichier CSV

print("Le nouveau dataset a été sauvegardé avec succès !")
```
